# ans9

```python
driver=webdriver.Chrome()
location_url='https://www.hostelworld.com/pwa/wds/s?q=London,%20England&country=London&city=London&type=city&id=3&from=2023-12-18&to=2023-12-21&guests=2&page=1'
driver.get(location_url)
```

```python
hostel_name=[]
distance_from_citycentre=[]
rating=[]
review=[]
price_from_dorms=[]
```

```python
names=driver.find_elements(By.XPATH,'//div[@class="property-name"]/span')
for i in names:
    sr=i.text
    hostel_name.append(sr)
```

```python
hostel_name
```

```python
distan=driver.find_elements(By.XPATH,'//div[@class="property-distance"]/span[2]')
for i in distan:
    ress=i.text
    distance_from_citycentre.append(ress)
```

```python
rate=driver.find_elements(By.XPATH,'//span[@class="number"]')
for i in rate:
    f=i.text
    rating.append(f)
```

```python
revew=driver.find_elements(By.XPATH,'//span[@class="keyword"]')
for i in revew:
    strin=i.text
    review.append(strin)
```

```python
rate=driver.find_elements(By.XPATH,'//strong[@class="current"]')
for i in rate:
    stri=i.text
    price_from_dorms.append(stri)
```

```python
dataframe=pd.DataFrame({'hostel_name':hostel_name,' distance_from_citycentre': distance_from_citycentre,
                       'ratings':rating,'reviews':review})
dataframe
```

```python
dataframe['price_from_dorms']=price_from_dorms
```

```python
dataframe
```
